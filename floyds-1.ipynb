```python
import numpy as np
import pandas as pd

def generate_adjacency_matrix(locations):
    num_locations = len(locations)
    np.random.seed(42)
    adj_matrix = np.random.randint(1, 100, size=(num_locations, num_locations))
    np.fill_diagonal(adj_matrix, 0)  # Distance to itself is 0
    return adj_matrix

def floyd_warshall(matrix):
    """Applies Floyd-Warshall algorithm to find shortest paths between all pairs."""
    n = len(matrix)
    dist = matrix.copy()
    
    for k in range(n):
        for i in range(n):
            for j in range(n):
                dist[i, j] = min(dist[i, j], dist[i, k] + dist[k, j])
    
    return dist

# Load dataset
file_path = "Chennai_neighbourhood_cleaned.csv"
df = pd.read_csv(file_path)
locations = df["Area"].tolist()

# Generate adjacency matrix
adj_matrix = generate_adjacency_matrix(locations)

# Compute shortest paths
shortest_paths = floyd_warshall(adj_matrix)

# Convert to DataFrame for better visualization
adj_matrix_df = pd.DataFrame(adj_matrix, index=locations, columns=locations)
shortest_paths_df = pd.DataFrame(shortest_paths, index=locations, columns=locations)

# Convert shortest paths DataFrame to array
shortest_paths_array = shortest_paths_df.values

# Initialize lists for features (X) and labels (y)
X = []
y = []

# Process the matrix to extract features and labels
for i in range(shortest_paths_array.shape[0]):
    for j in range(shortest_paths_array.shape[1]):
        if i != j and not np.isinf(shortest_paths_array[i][j]):  # Exclude diagonal and inf values
            X.append(shortest_paths_array[i][j])
            y.append(1 if shortest_paths_array[i][j] < 10 else 0)  # Label based on threshold

# Convert lists to arrays and reshape X for ML model
X = np.array(X).reshape(-1, 1)
y = np.array(y)

# Display the matrices

print("\nShortest Paths Matrix:")
print(shortest_paths_df)

# Display sample of features and labels
print("\nSample Features and Labels:")
print(list(zip(X.flatten()[:10], y[:10])))
```

```python
# Split into training and testing datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train a RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)

def print_solution(matrix, locations):
    """Prints the shortest paths matrix with location names."""
    print("\nShortest Paths Matrix:")
    df = pd.DataFrame(matrix, index=locations, columns=locations)
    print(df)

# Call the function correctly
print_solution(shortest_paths_array, locations)
```

```python
import folium

def predict_shortest_path(start_location, end_location):
    """Predicts and displays the shortest path for given locations on a map with a connecting line."""
    if start_location not in locations or end_location not in locations:
        print("Error: One or both locations not found in dataset.")
        return

    # Get indices for locations
    start_idx = locations.index(start_location)
    end_idx = locations.index(end_location)

    # Get the shortest distance from the precomputed matrix
    shortest_distance = shortest_paths_array[start_idx][end_idx]

    # Prepare input for the model
    input_feature = np.array([[shortest_distance]])

    # Predict the classification (short or long path)
    prediction = model.predict(input_feature)[0]
    prediction_label = "Short Path" if prediction == 1 else "Long Path"

    print(f"Predicted path classification: {prediction_label}")
    print(f"Shortest distance between {start_location} and {end_location}: {shortest_distance} units")

    # Get coordinates
    start_lat = df.loc[df["Area"] == start_location, "Latitude"].values[0]
    start_lon = df.loc[df["Area"] == start_location, "Longitude"].values[0]
    end_lat = df.loc[df["Area"] == end_location, "Latitude"].values[0]
    end_lon = df.loc[df["Area"] == end_location, "Longitude"].values[0]

    # Create map centered at the start location
    map_obj = folium.Map(location=[start_lat, start_lon], zoom_start=12)

    # Add markers for start and end locations
    folium.Marker(
        location=[start_lat, start_lon],
        popup=start_location,
        icon=folium.Icon(color="green", icon="play")
    ).add_to(map_obj)

    folium.Marker(
        location=[end_lat, end_lon],
        popup=end_location,
        icon=folium.Icon(color="red", icon="flag")
    ).add_to(map_obj)

    # Draw a line between the locations
    folium.PolyLine(
        locations=[[start_lat, start_lon], [end_lat, end_lon]],
        color="blue",
        weight=5,
        opacity=0.7
    ).add_to(map_obj)

    return map_obj

# Example usage:
start = "Egmore"
end = "Adyar"
map_output = predict_shortest_path(start, end)
map_output  # Displays the interactive map (works in Jupyter Notebook)
```
